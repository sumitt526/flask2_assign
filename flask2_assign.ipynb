GET and POST are two common HTTP request methods to retrieve or send data to the server.
GET uses url to send the data to the server and POST uses form/body to send the data to the server.
GET method is less secure than POST method as the data sent is the part of url and is visible to everyone while the POST method sent the data whole as a body therefore it is much secure.

# Qsn 3 solution:->

The redirect() function in Flask is used to send the user to a particular URL with a status code. It returns a response object that if called
redirects the client to the target location.

let's see it with a simple example:->

from flask import Flask, redirect, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return redirect(url_for('login'))

@app.route('/login')
def login():
    return 'Login Page'

if __name__ == '__main__':
    app.run()
    
In this example when a user accesses the root URL '/', they are redirected to the login page by calling redirect(url_for('login')).

Templates in Flask are files that contain static HTML and dynamic elements that are rendered by the server before being sent to the client. The render_template() function is used to render these templates. It takes the template’s filename as its first argument and a set of variables as the second argument. By default flask looks for templates in a folder called “templates”.

for example:-> render_template('index.html') this will render the 'index.html' page.
